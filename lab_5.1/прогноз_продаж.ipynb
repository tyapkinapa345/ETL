```python
# Загрузка данных
df_forecast = pd.read_csv("dubai_forecast.csv")

# Преобразуем столбец date в datetime для правильного порядка
df_forecast['date'] = pd.to_datetime(df_forecast['date'])

# Сортируем по дате (на случай, если порядок нарушен)
df_forecast.sort_values('date', inplace=True)

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(df_forecast['date'], df_forecast['temp_c'], marker='o', linestyle='-', color='orange')
plt.title('Температура в Дубае (прогноз на 3 дня)', fontsize=14)
plt.xlabel('Дата')
plt.ylabel('Температура (°C)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
